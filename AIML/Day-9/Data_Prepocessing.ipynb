```python
import pandas as pd
from sklearn.impute import SimpleImputer
data=pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
null_values=data.isnull().sum()
null_percentage = (data.isnull().sum() / len(data)) * 100
print("null values percentage: \n"+null_values.to_string())
print("null values percentage: \n"+null_percentage.to_string())
```

```python
imputer=SimpleImputer(strategy='most_frequent')
cols_to_impute = ['Age', 'Embarked']
data[cols_to_impute] = imputer.fit_transform(data[cols_to_impute])
data.drop('Cabin',axis=1,inplace=True)
```

```python
null_values=data.isnull().sum()
print(print("null values: \n"+null_values.to_string()))
```

```python
print(f"Shape of the DataFrame before removing duplicates:{data.shape}")
data.drop_duplicates(inplace=True)
print(f"Shape of the DataFrame after removing duplicates:{data.shape}")
```

```python
data = pd.get_dummies(data, columns=['Sex', 'Embarked'], drop_first=True)#both are nomimal->one hot encoding
```
